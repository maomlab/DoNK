```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ang5 = pd.read_csv('af_mpnn_data_5ang-0.3temp.csv', sep='\t')

ang10 = pd.read_csv('af_mpnn_data_10ang-0.3temp.csv', sep='\t')
```

```python
ang5.head()
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(ang5['seq_recovery'], ang5['plddt'], c=ang5['local_prob'], cmap='viridis', s=2, alpha=0.7)

# Add a colorbar
colorbar = plt.colorbar(scatter)
colorbar.set_label('proteinmpnn local score')

# Label axes and title
plt.xlabel('proteinmpnn sequence recovery')
plt.ylabel('alphafold2 pLDDT')
plt.title('5 angstrom 0.3 temperature designs')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(ang5['local_prob'], ang5['plddt'], c=ang5['seq_recovery'], cmap='viridis', s=2, alpha=0.7)

# Add a colorbar
colorbar = plt.colorbar(scatter)
colorbar.set_label('proteinmpnn sequence recovery')

# Label axes and title
plt.xlabel('proteinmpnn local score')
plt.ylabel('alphafold2 pLDDT')
plt.title('5 angstrom 0.3 temperature designs')

plt.show()
```
